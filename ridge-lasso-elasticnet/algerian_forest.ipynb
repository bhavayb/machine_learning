```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header = 1) ##take header = 1 means start from row 1
```

```python
dataset.head()
```

```python
dataset.info()
```

## data cleaning

```python
dataset.isnull().sum()
```

```python
dataset.isnull().any(axis=1)
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,"Region"] = 0
dataset.loc[122:, "Region"] = 1;
```

```python
df = dataset
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df[df.isnull().any(axis=1)]

```

```python
## removing the null values
df = df.dropna().reset_index(drop = True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in col names
df.columns = df.columns.str.strip()
df.columns
```

#### change the required columns as a int data type

```python
df[['month','day', 'year', 'Temperature','RH','Ws']] = df[['month','day', 'year', 'Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
## changing the other col to float data type
strings = [features for features in df.columns if df[features].dtypes=='str']
```

```python
for i in strings:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
## lets save the clean dataset
df.to_csv('Algerian_forest_fires_dataset_cleaned_dataset.csv', index= False)
```
